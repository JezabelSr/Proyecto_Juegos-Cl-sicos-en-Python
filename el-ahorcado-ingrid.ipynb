# PRIMER CODIGO CON UNA SOLA PALABRA SECRETA
- PENDIENTE: DAR PISTA DE LA PALABRA (PRINT CON 4 _ _ _ _)
- IMPRIMIR CADA LETRA CUANDO NO HAY ACIERTO
- IMPRIMIR LA PALABRA QUE ERA CUANDO NO HAY ACIERTO. (print de la palabra secreta era:)
- INCLUIRE MOTICONOS Y COLORES


```python
## VARIABLES A TENER EN CUENTA##

palabra_secreta = "gato"                         # 1. Palabra secreta (puede ser import random)
tablero = ["_"] * len(palabra_secreta)           # 2. Crear el tablero * num de letras
letras_usadas = []                              # 3. almacenar las letras usadas por el jugador
intentos = 6 

while intentos > 0:
    letra = input("Bienvenida al juego El Ahorcado! Di una letra:")
    if letra in letras_usadas:
        print("Esa letra ya la usaste....Di otra letra.")
        continue    
    if letra in palabra_secreta:
        print("¡Bien! La letra está en la palabra")
        letras_usadas.append(letra)
        for i in range(len(palabra_secreta)):
            if palabra_secreta[i] == letra:
                tablero[i] = letra
        print(tablero)
    if letra not in palabra_secreta:
        intentos -= 1 
        print("¡has fallado! vuelve a intentarlo:")
        print(f"Te quedan {intentos} intentos")
    if "_" not in tablero:
        print("¡Bien hecho, has ganado la partida!!")
        break
if intentos == 0:
    print("Lo siento has consumido todos tus intentos..¡has perdido!")  

```

# SEGUNDO CODIGO CON RANDOM:
PENDIENTE:
- IMPRIMIR CADA LETRA CUANDO NO HAY ACIERTO
- IMPRIMIR LA PALABRA QUE ERA CUANDO NO HAY ACiERTO. (print de: la palabra secreta era:)

```python
## VARIABLES A TENER EN CUENTA##

import random
palabras = [
    "python",
    "variable",
    "funcion",
    "bucle",
    "condicion",
    "lista",
    "diccionario",
    "programa",
    "algoritmo",
    "computadora"
]


palabra_secreta = random.choice(palabras)               # 1. Palabra secreta (import random)
tablero = ["_"] * len(palabra_secreta)                  # 2. Crear el tablero * num de letras
letras_usadas = []                                       # 3. almacenar las letras usadas por el jugador
intentos = 6 

while intentos > 0:
    letra = input("Bienvenida al juego El Ahorcado! Di una letra:")
    if len(letra) > 1:
        print("¡Error! Debes introducir una sola letra:")
        continue
    if letra in letras_usadas:
        print("Esa letra ya la usaste....Di otra letra.")
        continue    
    if letra in palabra_secreta:
        print("¡Bien! La letra está en la palabra")
        letras_usadas.append(letra)
        for i in range(len(palabra_secreta)):
            if palabra_secreta[i] == letra:
                tablero[i] = letra
        print(tablero)
    if letra not in palabra_secreta:
        intentos -= 1 
        print("¡has fallado! vuelve a intentarlo:")
        print(f"Te quedan {intentos} intentos")
    if "_" not in tablero:
        print("¡Bien hecho, has ganado la partida!!")
        break
if intentos == 0:
    print("Lo siento has consumido todos tus intentos..¡has perdido!")  
```
